```python
#Importar librerias Pandas
import pandas as pd
```

```python
#Leer archivo csv
df = pd.read_csv("C:/Users/Lenovo/Desktop/Boot Camp/Python/Desafio_5/Apoyo Desafío - Tipos y Estructuras de Datos (Parte II)/resumen_resultados_astronautas.csv")

```

```python
df.head()
```

```python
#Crea la columna "IMC"
df["IMC"] = df.peso/((df.altura/100)*(df.altura/100))
```

```python
#Crea la columna "promedio evaluaciones"
df["promedio_evaluaciones"] = (df.evaluacion_fisica+df.evaluacion_psicologica+df.evaluacion_adaptativa+df.evaluacion_medica)/4
```

```python
df.head(1)
```

```python
"Obtener IMC está entre 18.5 y 24.9"
df_mod= df.query("IMC <= 24.9 and IMC >= 18.5")
df_mod
```

```python
#Agregar condición de promedio de las 4 evaluaciones
```

```python
df_mod_2= df_mod.query("promedio_evaluaciones >= 87")
df_mod_2
```

```python
#Reinicia los indices
df3 = df_mod_2.reset_index(drop=True)
df3
```

```python
#Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del DataFrame resultante.

df3['califica'] = df3['califica'].replace('No','Si')
df3
```

```python
#Exporta el DataFrame resultante a un archivo .csv separado por comas con el nombre "astronautas_calificados.csv"


df3.to_csv("astronautas_calificados.csv", index = False, sep = ',', header = True)
```
