```python
#Cruza todos los DataFrames usando validación many_to_one, y guarda el resultado en una nueva variable llamada df_base.
df_merged = SalesOrdersSheets.merge(CustomersSheet,on = '_CustomerID',how ='left', validate = 'many_to_one')
```

```python
df_merged1 = df_merged.merge(StoreLocationsSheet,on = '_StoreID',how ='left', validate = 'many_to_one')
```

```python
df_merged2 = df_merged1.merge(ProductsSheet,on = '_ProductID',how ='left', validate = 'many_to_one')
```

```python
df_base = df_merged2.merge(SalesTeamSheet,on = '_SalesTeamID',how ='left', validate = 'many_to_one')
```

```python
df_base.head(1)
```

```python
##Agrega las siguientes columnas a df_ base:ProcurementDays, ShippingDays, DeliveryDays y CustomerDays

df_base['ProcurementDays'] = pd.to_datetime(df_base['OrderDate']) - pd.to_datetime(df_base['ProcuredDate'])
```

```python
df_base.shape
```

```python
df_base['ShippingDays'] = pd.to_datetime(df_base['ShipDate']) - pd.to_datetime(df_base['OrderDate'])
```

```python
df_base.shape
```

```python
df_base['DeliveryDays'] = pd.to_datetime(df_base['DeliveryDate']) - pd.to_datetime(df_base['ShipDate'])
```

```python
df_base.shape
```

```python
df_base['CustomerDays'] = df_base['ShippingDays'] + df_base['DeliveryDays']
```

```python
df_base.shape
```

```python
df_base.head(1)
```

```python
#Agrega una nueva columna discreta al df_base, con el nombre CustomerDaysInterval que clasifique los valores de la columna CustomerDays en los siguientes intervalos:

intervalos = [0, 15, 30, 45, 60, 75, 90] 
            
```

```python
etiquetas = ['0 to 15 days','15 to 30 days', '30 to 45 days', '45 to 60 days', '60 to 75 days', '75 to 90 days']
```

```python
df_base['CustomerDaysInterval'] = pd.cut(df_base['CustomerDays'].dt.days, bins=intervalos, labels=etiquetas, right = True)
```

```python
df_base.head(1)
```

```python
# Utiliza una tabla pivote para mostrar el conteo de órdenes agrupando por Sales Team en las filas y CustomerDaysInterval en las columnas.
df_base_reporte1 = df_base.pivot_table(
                                  index = "Sales Team"
                                ,columns = "CustomerDaysInterval"
                                 , values = "OrderNumber"
                                 , aggfunc= "count" 
                                 , margins = True
                                            )

df_base_reporte1 
```

```python
import numpy as np
df_base_reporte1 .replace(np.nan, 0)
```

```python
#Agrega la siguiente columna calculada a la base:GrossMargin
df_base['GrossMargin'] = pd.to_numeric(df_base['Order Quantity']*(df_base['Unit Price']*(1-df_base['Discount Applied']) - df_base['Unit Cost']))
df_base.head(5)
                                                                                                     
```

```python
#En esta parte debes aplicar un porcentaje a la columna GrossMargin, pero ese porcentaje dependerá del intervalo en que se encuentre GrossMargin.

intervalos_1 = [0, 100, 1000, 10000,100000] 
```

```python
etiquetas_1 = pd.to_numeric([0.05,0.1, 0.15, 0.2])
```

```python
df_base['CommissionsPercentage'] = pd.to_numeric(pd.cut(df_base['GrossMargin'], bins=intervalos_1, labels=etiquetas_1, right = True))
```

```python
df_base.head(1)
```

```python
#Calcula el monto de la comisión, utilizando la Tabla del Anexo.

df_base['CommissionsAmount'] = pd.to_numeric(df_base['GrossMargin']*df_base['CommissionsPercentage'])
df_base.head(5)
                
```

```python
#Calcula la comisión sobre el margen bruto.

df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']
df_base.head(5)
                
```

```python
#Utilizando una tabla pivote, muestra la suma de los valores GrossMargin, NetMargin y CommissionAmount agrupando por Sales Team en las filas

df_base_reporte2 = df_base.pivot_table(
                                  index = "Sales Team"
                                  ,values = ("CommissionsAmount", "GrossMargin", "NetMargin")
                                  , aggfunc= "sum" 
                                  , margins = True
                                        )

df_base_reporte2
```
