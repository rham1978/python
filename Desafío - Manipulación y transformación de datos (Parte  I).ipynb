```python
#Carga los datos y crea un DataFrame con ellos.
df_csv= pd.read_csv("C:/Users/Lenovo/Desktop/Boot Camp/Python/Desafio_6/Apoyo Desafío - Manipulación y transformación de datos (Parte  I)/Cleaned_DS_Jobs.csv")

df_csv
```

```python
#Utiliza la siguiente lista de valores que serán considerados como nulos:
import numpy as np
```

```python
df_csv1 = df_csv.replace(["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"],np.nan )
df_csv1.head(1)
```

```python
#Elimina todas las filas con datos faltantes

df_csv1.dropna(axis = 0).reset_index(drop=True)
```

```python
#A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo y Máximo.
df_csv1[['Salario Estimado Mínimo','Salario Estimado Máximo']]=df_csv1['Salary Estimate'].str.split('-',expand=True).apply(lambda x: x.str.strip()).reset_index(drop=True)
df_csv1
```

```python
#Realiza la recodificación de la columna Size con los valores de la siguiente tabla:
df_replaced=df_csv1.replace({ 'Size': { '10000+ employees': 'Mega Empresas',
                            '5001 to 10000 employees': 'Grandes Empresas',
                            '1001 to 5000 employees': 'Medianas Empresas',
                            '201 to 500 employees': 'Pequeñas Empresas',
                            '51 to 200 employees': 'Pequeñas Grandes Empresas',
                            '501 to 1000 employees': 'Microempresas',
                             'Unknown':'Empresas sin Información'}})               
                         
df_replaced                                              
```

```python
#Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y la media del salario estimado máximo por tamaño de empresa.
df_csv1_pivot_reporte = pd.pivot_table(
                                   data = df_replaced 
                                 #,columns = ("Salario Estimado Mínimo", "Salario Estimado Máximo")
                                 , index = "Size"
                                 , values = ("Salario Estimado Mínimo", "Salario Estimado Máximo")
                                 , aggfunc= "median"
                                 , margins = True
)

df_csv1_pivot_reporte

```
